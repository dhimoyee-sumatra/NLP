```python
with open ("emma_tokens1.txt") as f:
    text= f.read()
text
```

```python
text=text.strip().split('\n')
text
```

```python
dict= []
for x in range(1, len(text)):
    str= text[x-1] +' ' + text[x]
    dict.append(str)
dict
```

```python
from collections import Counter 
cnt1 = Counter()
for word in text:
    cnt1[word] += 1
print (cnt1)
```

```python
cnt2 = Counter()
for word in dict:
    cnt2[word] += 1
cnt2['<s> she']
```

```python
prob1={}
prob2={}
for y in text:
    prob2[y]={}
for x in range(0, len(dict)):
    str= dict[x].split()
    prob1[dict[x]]=round(cnt2[dict[x]]/cnt1[str[0]],2)
    var= {prob1[dict[x]]: str[1]}
    #prob2[str[0]]={}
    prob2[str[0]].update(var)
    #prob2[str[0]]={prob1[dict[x]]: str[1]}
prob2    
```

```python
import random
man= '<s>'
list= prob2[man]
for key, value in list.items():
    print (key, value)
```

```python
for x in range(0, 10):
    man= '<s>'
    sun= '<s>'
    while not sun.endswith('<e>'):
        list= prob2[man]
        var= random.random()
        for key, value in list.items():
            if var < key:
                sun+=value
                man= value
            else:
                sun+=value
                man=value
    print(sun)
```
